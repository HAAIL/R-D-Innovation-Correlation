```python
import re
import matplotlib.pyplot as plt
import requests

filename2018 = "microsoft_response2018.txt"
filename2019 = "microsoft_response2019.txt"
filename2020 = "microsoft_response2020.txt"
filename2021 = "microsoft_response2021.txt"
filename2022 = "microsoft_response2022.txt"
full_dict = {}
```

```python
#year 2018
with open(filename2018, 'r') as file:
    text = file.read()

# Extracting the topics from the text
topics = re.findall(r'\d+\. (.+)', text)

# Counting the occurrences of each topic
topic_count = {}
for topic in topics:
    topic_count[topic] = topic_count.get(topic, 0) + 1

# Displaying the dictionary
new_dict = {}
print("Total topics:", len(topic_count))
for key, val in topic_count.items():
    if val > 1:
        print(key, ": ", val)
        new_dict[key] = val
# print(topic_count)

full_dict["2018"] = new_dict
```

```python
# Extract topic names and their corresponding counts
topic_names = list(new_dict.keys())
topic_values = list(new_dict.values())

# Create the bar chart
plt.figure(figsize=(12, 10))
plt.barh(topic_names, topic_values, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Topics')
plt.title('Topic Counts')
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
# Calculate total counts for all topics in each year
year_totals = {}
for year, topics_dict in full_dict.items():
    total_count = sum(topics_dict.values())
    year_totals[year] = total_count

# Extract year and total count as separate lists
years = list(year_totals.keys())
total_counts = list(year_totals.values())

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, total_counts, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.title('Total Count for Topics by Year')
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
import numpy as np

# Sample dictionary (replace this with your actual data)
data = full_dict

# Extract years and topics
years = list(data.keys())
topics = list(set(topic for topics_dict in data.values() for topic in topics_dict.keys()))

# Create a matrix to store the counts for each topic in each year
topic_counts_matrix = np.zeros((len(years), len(topics)))

# Fill the matrix with topic counts
for i, year in enumerate(years):
    for j, topic in enumerate(topics):
        topic_counts_matrix[i, j] = data[year].get(topic, 0)

# Create the grouped bar chart
plt.figure(figsize=(14, 6))  # Increase the figure size
bar_width = 0.15  # Adjust the bar width
bar_positions = np.arange(len(years))

for i, topic in enumerate(topics):
    plt.bar(bar_positions + (i * bar_width), topic_counts_matrix[:, i], bar_width, label=topic)

plt.xlabel('Year')
plt.ylabel('Topic Counts')
plt.title('Topic Counts by Year and Topic')
plt.xticks(bar_positions + (bar_width * (len(topics) - 1) / 2), years, rotation=45, ha='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Topics')
plt.tight_layout()

# Add more space between different years
plt.subplots_adjust(bottom=0.2)

# Display the grouped bar chart
plt.show()
```
