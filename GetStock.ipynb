```python
def get_historical_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

def plot_stock_data(stock_data, symbol):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data['Close'], label=f'{symbol} Close Price', color='blue')
    plt.title(f'{symbol} Historical Stock Data')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    symbol = 'MSFT'
    end_date = datetime.date.today()
    start_date = end_date - datetime.timedelta(days=365*5)

    stock_data = get_historical_data(symbol, start_date, end_date)
    if not stock_data.empty:
        print(f'{symbol} Historical Stock Data:')
        print(stock_data)

        plot_stock_data(stock_data, symbol)
    else:
        print(f"Failed to retrieve historical stock data for {symbol}.")
```
